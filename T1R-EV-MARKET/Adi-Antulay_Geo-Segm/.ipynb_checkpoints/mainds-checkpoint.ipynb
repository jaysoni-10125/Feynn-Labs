```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel('finaldataset.xlsx', sheet_name='Sheet1')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Extract numeric data for clustering
numeric_columns = ['CNG ONLY', 'PETROL/CNG', 'DIESEL', 'ELECTRIC(BOV)', 'PETROL']
data_numeric = df[numeric_columns]
```

Why: Standardizing the data ensures all features contribute equally to the clustering process, preventing features with larger ranges from dominating.


```python
# Standardize the data for clustering
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_numeric)
```

```python
# Apply K-Means Clustering
print("\nApplying K-Means Clustering:")
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(data_scaled)
```

```python
# Visualize clustering results with PCA
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_scaled)
df['PCA1'] = data_pca[:, 0]
df['PCA2'] = data_pca[:, 1]
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df, palette='viridis')
plt.title('K-Means Clustering Visualization')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.legend(title='Cluster')
plt.show()

```

 Why: K-Means helps segment the data into distinct groups. Visualizing these clusters using PCA provides an intuitive understanding of the results.


```python
# Aggregate data by State
state_summary = df.groupby('State').sum(numeric_only=True)
```

```python
# Normalize data for comparison
state_normalized = state_summary.div(state_summary.sum(axis=1), axis=0)
```

Why: Aggregating and normalizing data by state allows us to compare distributions across different regions effectively.



```python
# Visualization 1: Distribution of Vehicle Classes by Fuel Type
plt.figure(figsize=(10, 6))
fuel_columns = ['CNG ONLY', 'PETROL/CNG', 'DIESEL', 'ELECTRIC(BOV)', 'PETROL']
state_summary[fuel_columns].sum().plot(kind='bar', color=['green', 'orange', 'blue', 'purple', 'red'])
plt.title('Total Distribution of Vehicle Classes by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45)
plt.show()
```

Why: This visualization highlights the overall fuel type preferences, helping identify market trends and opportunities.


```python
# Visualization 2: Heatmap of Normalized Data by State and Fuel Type
plt.figure(figsize=(12, 8))
sns.heatmap(state_normalized, annot=False, cmap='viridis')
plt.title('Normalized Vehicle Distribution by State and Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('State')
plt.show()
```

Why: The heatmap reveals state-level differences in vehicle fuel distributions, aiding in region-specific decision-making.


```python
# Segmentation: Identify States with the highest number of Electric Vehicles
electric_vehicle_data = state_summary['ELECTRIC(BOV)']
print("\nTop 5 States with the highest number of Electric Vehicles:")
print(electric_vehicle_data.sort_values(ascending=False).head())
```

Why: Identifying top states for electric vehicles can guide marketing and infrastructure development efforts.


```python
# Silhouette Score for K-Means
silhouette_kmeans = silhouette_score(data_scaled, kmeans.labels_)
print(f"Silhouette Score for K-Means: {silhouette_kmeans}")
```

 Why: Comparing clustering methods (e.g., K-Means and DBSCAN) ensures robustness and helps identify the best approach for segmentation.


```python
# DBSCAN Clustering
print("\nApplying DBSCAN Clustering:")
dbscan = DBSCAN(eps=1.5, min_samples=5)
df['DBSCAN_Cluster'] = dbscan.fit_predict(data_scaled)


# Visualization of DBSCAN results
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='DBSCAN_Cluster', data=df, palette='viridis')
plt.title('DBSCAN Clustering Visualization')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.legend(title='Cluster')
plt.show()
```

 Why: DBSCAN helps identify clusters with varying densities, complementing K-Means' fixed cluster approach.

Visualizations like bar charts and heatmaps offered insights into fuel type trends and regional distributions. The results can guide decision-making in market targeting and resource allocation.
