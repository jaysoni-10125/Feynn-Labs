```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
# Load the datasets
file_path1 = r'C:\Users\lenovo\Desktop\p3\dataset2\CS_Sanctioned.csv'
file_path2 = r'C:\Users\lenovo\Desktop\p3\dataset2\state_wise_gov.csv'
file_path3 = r'C:\Users\lenovo\Desktop\p3\dataset2\CS_10-12-21.csv'
file_path4 = r'C:\Users\lenovo\Desktop\p3\dataset2\CS_Highway.csv'

data1 = pd.read_csv(file_path1)
data2 = pd.read_csv(file_path2)
data3 = pd.read_csv(file_path3)
data4 = pd.read_csv(file_path4)
```

```python
# Remove rows labeled as "Total" from all datasets
data1 = data1[~data1['State/UT-wise'].str.contains('Total', case=False, na=False)]
data2 = data2[~data2['State/UT'].str.contains('Total', case=False, na=False)]
data3 = data3[~data3['State/ UT'].str.contains('Total', case=False, na=False)]
data4 = data4[~data4['Highways/Expressways'].str.contains('Total', case=False, na=False)]
```

```python
# Step 1: Analyze Each Dataset Separately
# Dataset 1: CS_Sanctioned
plt.figure(figsize=(12, 6))
sns.barplot(x='State/UT-wise', y='Number of Electric Vehicle Charging Sanctioned', data=data1)
plt.xticks(rotation=90)
plt.title('EV Charging Stations Sanctioned by State')
plt.ylabel('Number of Stations')
plt.xlabel('State/UT')
plt.tight_layout()
plt.savefig('sanctioned_ev_stations.png')
plt.show()
```

```python
# Dataset 2: state_wise_gov
plt.figure(figsize=(12, 6))
data2_sorted = data2.sort_values(by='Grand Total', ascending=False)
sns.barplot(x='State/UT', y='Grand Total', data=data2_sorted)
plt.xticks(rotation=90)
plt.title('State-wise Government Activities (Grand Total)')
plt.ylabel('Grand Total Activities')
plt.xlabel('State/UT')
plt.tight_layout()
plt.savefig('state_wise_gov_activities.png')
plt.show()
```

```python
# Dataset 3: CS_10-12-21
plt.figure(figsize=(12, 6))
sns.barplot(x='State/ UT', y='Charging Stations', data=data3)
plt.xticks(rotation=90)
plt.title('Charging Stations by State (10-12-21)')
plt.ylabel('Number of Charging Stations')
plt.xlabel('State/UT')
plt.tight_layout()
plt.savefig('state_charging_stations.png')
plt.show()
```

```python
# Dataset 4: CS_Highway
plt.figure(figsize=(12, 6))
data4_sorted = data4.sort_values(by='Charging Stations', ascending=False)
sns.barplot(x='Highways/Expressways', y='Charging Stations', data=data4_sorted)
plt.xticks(rotation=90)
plt.title('Charging Stations on Highways/Expressways')
plt.ylabel('Number of Charging Stations')
plt.xlabel('Highway/Expressway')
plt.tight_layout()
plt.savefig('highway_charging_stations.png')
plt.show()
```

```python
# Step 2: Market Segmentation

# Dataset 1: Prioritize states with high numbers of sanctioned stations
priority_states = data1.sort_values(by='Number of Electric Vehicle Charging Sanctioned', ascending=False).head(5)
print("Priority States for EV Station Investment:")
print(priority_states)
```

```python
# Dataset 2: Identify states with high government activity levels
high_activity_states = data2_sorted.head(5)
print("States with Highest Government Activities:")
print(high_activity_states)
```

```python
# Dataset 3: Analyze state-level charging station distribution
high_charging_states = data3.sort_values(by='Charging Stations', ascending=False).head(5)
print("States with Highest Charging Stations:")
print(high_charging_states)
```

```python
# Dataset 4: Highlight highways with the most charging stations
priority_highways = data4_sorted.head(5)
print("Highways with Highest Charging Stations:")
print(priority_highways)
```

```python
# Clustering Analysis: State segmentation based on metrics
state_metrics = data1[['State/UT-wise', 'Number of Electric Vehicle Charging Sanctioned']].merge(
    data2[['State/UT', 'Grand Total']], left_on='State/UT-wise', right_on='State/UT', how='inner')
state_metrics = state_metrics.rename(columns={'Number of Electric Vehicle Charging Sanctioned': 'Sanctioned Stations', 'Grand Total': 'Government Activity'})

kmeans = KMeans(n_clusters=3, random_state=42)
state_metrics['Cluster'] = kmeans.fit_predict(state_metrics[['Sanctioned Stations', 'Government Activity']])

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Sanctioned Stations', y='Government Activity', hue='Cluster', data=state_metrics, palette='Set2')
plt.title('State Clusters Based on Sanctioned Stations and Government Activity')
plt.xlabel('Sanctioned Stations')
plt.ylabel('Government Activity')
plt.legend(title='Cluster')
plt.tight_layout()
plt.savefig('state_clusters.png')
plt.show()
```
